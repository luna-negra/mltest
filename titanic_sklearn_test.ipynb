### Some columns have Null value

### fillna: Cabin

```python
data.fillna({"Cabin": "N"}, inplace=True)
data["Cabin"] = data["Cabin"].apply(lambda x: x[0])
```

### fillna(): Embarked

```python
print(data["Embarked"].value_counts())
data["Embarked"].isnull().value_counts()
```

```python
data.fillna({"Embarked": "N"}, inplace=True)
data
```

```python
data.isnull().value_counts()
# There is no N/A Value.
```

```python
# get rid of unneccesary columns: PassengerId, Name, Ticket, Cabin
new_data = data.drop(["Name", "Ticket", "Cabin"], axis=1)
new_data
```

```python
# Labeling for Model Learning
print(new_data.info())

# Label Encoding for "Sex" and "Embarked"
from sklearn.preprocessing import LabelEncoder

features = ["Sex", "Embarked"]
for f in features:
    encoder = LabelEncoder()
    encoder.fit(new_data[f])
    new_data[f] = encoder.transform(new_data[f])

new_data
```

### Analysis the Passenger

```python
# split data in train.csv into test and train data set
# Data will be learned by DecisionTree, LogisticRegression and Random Forest.

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


# split data set
X_data = new_data.drop("Survived", axis=1)
Y_data = new_data["Survived"]
X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_data, test_size=0.2, random_state=12)
```

```python
dt_model = DecisionTreeClassifier()
lr_model = LogisticRegression()
rf_model = RandomForestClassifier()

model_list = [dt_model, lr_model, rf_model]
accuracy = []

for model in model_list:
    model.fit(X_train, Y_train)
    pred = model.predict(X_test)
    accuracy.append(accuracy_score(pred, Y_test))
```

```python
import numpy as np

result = {}
print(f"!! Normal Test Train Split")

for n in range(len(model_list)):
    result.update({model_list[n]: np.round(accuracy[n], 2)})

print(result)
```

### Each model has not high prediction score.
- Possibility of Overfit

=> Use Cross Validation with KFold and StratifiedKFold

#### KFold

```python
from sklearn.model_selection import KFold


kfold = KFold(n_splits=5)
max_score_model = ""
max_score = 0

for model in model_list:    
    for train_index, test_index in kfold.split(X_data):
        X_train, X_test = X_data.iloc[train_index], X_data.iloc[test_index]
        Y_train, Y_test = Y_data.iloc[train_index], Y_data.iloc[test_index]

        model.fit(X_train, Y_train)
        accuracy = np.round(accuracy_score(model.predict(X_test), Y_test), 2)
        if max_score < accuracy:
            max_score_model = str(model.__class__).split(".")[-1][:-2]
            max_score = accuracy

print(max_score_model)
print(max_score)
```

#### StratifiedKFold with cross_val_score()


```python
from sklearn.model_selection import cross_val_score


max_score_model = ""
max_score = 0

for model in model_list:
    scores = cross_val_score(model, X_data, Y_data, cv=5)

    for score in scores:
        if max_score < score:
            max_score_model = str(model.__class__).split(".")[-1][:-2]
            max_score = np.around(score, 2)

print(max_score_model)
print(max_score)
```

### Random Forest predict the test data with highest score.
- do tuning on hyper parameters with GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

model = DecisionTreeClassifier()

# check parameters
print(model.get_params())
```

```python
params = {"max_depth": range(10), "max_leaf_nodes": range(10), "min_samples_leaf": range(10)}

# Define GridSearchCV
gscv = GridSearchCV(model, param_grid=params, scoring="accuracy", cv=5)
gscv.fit(X_train, Y_train)
```

```python
# Check the best Params
print(gscv.best_params_)
```

```python
# Check the best score
print("---")
print(f"GridSearchCV Best Score: {np.round(gscv.best_score_, 2)}")
print("---")
```

```python
# Get the best Param and apply
tuned_model = gscv.best_estimator_
pred = tuned_model.predict(X_test)
print("---")
print(f"Model with Best HyperParameters Prediction: {np.around(accuracy_score(pred, Y_test), 2)}")
print("---")
```
